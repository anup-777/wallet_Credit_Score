```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load the JSON file
#df = pd.read_json('user-wallet-transactions.json')

# Convert timestamp to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Sort by user and timestamp
df = df.sort_values(by=['userWallet', 'timestamp'])

# Feature engineering per userWallet
def extract_wallet_features(group):
    group = group.sort_values('timestamp')
    time_diffs = group['timestamp'].diff().dt.total_seconds().dropna() / 3600  # in hours

    days_active = group['timestamp'].dt.date.nunique()

    return pd.Series({
        'deposit_count': (group['action'] == 'deposit').sum(),
        'borrow_count': (group['action'] == 'borrow').sum(),
        'repay_count': (group['action'] == 'repay').sum(),
        'redeem_count': (group['action'] == 'redeemunderlying').sum(),
        'liquidation_count': (group['action'] == 'liquidationcall').sum(),
        'total_txn_count': len(group),
        'unique_days_active': days_active,
        'avg_time_between_txn': time_diffs.mean() if not time_diffs.empty else 0,
        'txn_freq_per_day': len(group) / days_active if days_active > 0 else 0
    })

wallet_features = df.groupby('userWallet').apply(extract_wallet_features).reset_index()

# Fill any NaNs
wallet_features.fillna(0, inplace=True)

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(wallet_features.drop(columns=['userWallet']))

# Clustering using KMeans
kmeans = KMeans(n_clusters=5, random_state=42)
wallet_features['cluster'] = kmeans.fit_predict(X_scaled)

# Assign credit score based on clusters (you can customize this mapping)
cluster_score_map = {
    0: 800,
    1: 400,
    2: 1000,
    3: 200,
    4: 600
}

wallet_features['credit_score'] = wallet_features['cluster'].map(cluster_score_map)

# Final output
wallet_scores = wallet_features[['userWallet', 'credit_score']]
wallet_scores.to_csv('wallet_credit_scores.csv', index=False)
print(wallet_scores)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load the scored wallet data
df = pd.read_csv('wallet_credit_scores.csv')

# Define score bins and labels
bins = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
labels = ['0â€“100', '100â€“200', '200â€“300', '300â€“400', '400â€“500',
          '500â€“600', '600â€“700', '700â€“800', '800â€“900', '900â€“1000']

# Bin the credit scores
df['score_range'] = pd.cut(df['credit_score'], bins=bins, labels=labels, right=False)

# Count how many wallets fall into each range
score_counts = df['score_range'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 6))
bars = plt.bar(score_counts.index, score_counts.values, color='skyblue', edgecolor='black')

plt.title('Wallet Credit Score Distribution')
plt.xlabel('Score Range')
plt.ylabel('Number of Wallets')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate bars with counts
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 5, str(int(height)),
             ha='center', va='bottom', fontsize=10)

# Ensure output directory exists
os.makedirs('outputs', exist_ok=True)

# Save plot
plt.tight_layout()
plt.show()
plt.savefig('outputs/score_distribution.png')
plt.close()
```
